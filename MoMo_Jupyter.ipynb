# Data processing

### Task 1

#### Calculate Loyalty Points

```python
df = pd.read_excel("Test.xlsx", sheet_name="Data. Transactions")
```

```python
df.head()
```

```python
for i in range(len(df)):
    if df["Service Group"][i] ==  "supermarket":
        df["Loyalty Points"][i] = (df["GMV"][i] * 1) / 1000
    elif df["Service Group"][i] ==  "data":
        df["Loyalty Points"][i] = (df["GMV"][i] * 10) / 1000
    elif df["Service Group"][i] ==  "cvs":
        df["Loyalty Points"][i] = (df["GMV"][i] * 1) / 1000
    elif df["Service Group"][i] ==  "marketplace":
        df["Loyalty Points"][i] = (df["GMV"][i] * 1) / 1000
    elif df["Service Group"][i] ==  "Coffee chains and Milk tea":
        df["Loyalty Points"][i] = (df["GMV"][i] * 1) / 1000
    elif df["Service Group"][i] ==  "Offline Beverage":
        df["Loyalty Points"][i] = (df["GMV"][i] * 1) / 1000
```

```python
df["Loyalty Points"] = df["Loyalty Points"].astype("int")
```

```python
df.head()
```

```python
df.to_excel("Final.xlsx", sheet_name= "Data. Transactions")
```

#### Identify User's Rank Name

```python
df1 = pd.read_excel("Test.xlsx", sheet_name="Loyalty Ranking")
```

```python
df1 = df1.drop_duplicates().groupby("User_id").sum().reset_index()
```

```python
df1.head()
```

```python
for i in range(len(df1)):
    if df1["Calculated_points"][i] >= 1 and df1["Calculated_points"][i] <= 999:
        df1["Rank_name"][i] = "STANDARD"
    elif df1["Calculated_points"][i] >= 1000 and df1["Calculated_points"][i] <= 1999:
        df1["Rank_name"][i] = "SILVER"
    elif df1["Calculated_points"][i] >= 2000 and df1["Calculated_points"][i] <= 4999:
        df1["Rank_name"][i] = "GOLD"
    else:
        df1["Rank_name"][i] = "DIAMOND"
    
```

```python
df1.head()
```

```python
df1["Rank_name"].value_counts()
```

```python
df1.to_excel("Final1.xlsx", sheet_name= "Loyalty Ranking")
```

#### Calculate cashback cost for each Transaction

```python
for i in range(len(df)):
    #if df["Rank_name"][i] == "STANDARD":
    #    df.cashback = 0
    if df["Rank_name"][i] == "SILVER":
        if df["Service Group"][i] == "cvs":
            df["$cashback"][i] = df["GMV"][i] * 0.05
        elif df["Service Group"][i] == "Offline Beverage":
            df["$cashback"][i] = df["GMV"][i] * 0.05
        elif df["Service Group"][i] == "data":
            df["$cashback"][i] = df["GMV"][i] * 0.05
        else:
            df["$cashback"][i] = 0
    elif df["Rank_name"][i] == "GOLD":
        if df["Service Group"][i] == "cvs":
            df["$cashback"][i] = df["GMV"][i] * 0.07
        elif df["Service Group"][i] == "Offline Beverage":
            df["$cashback"][i] = df["GMV"][i] * 0.07
        elif df["Service Group"][i] == "data":
            df["$cashback"][i] = df["GMV"][i] * 0.07
        elif df["Service Group"][i] == "marketplace":
            df["$cashback"][i] = df["GMV"][i] * 0.02
        elif df["Service Group"][i] == "supermarket":
            df["$cashback"][i] = df["GMV"][i] * 0.02
        else:
            df["$cashback"][i] = 0
    elif df["Rank_name"][i] == "DIAMOND":
        if df["Service Group"][i] == "cvs":
            df["$cashback"][i] = df["GMV"][i] * 0.12
        if df["Service Group"][i] == "Offline Beverage":
            df["$cashback"][i] = df["GMV"][i] * 0.12
        if df["Service Group"][i] == "data":
            df["$cashback"][i] = df["GMV"][i] * 0.12
        if df["Service Group"][i] == "marketplace":
            df["$cashback"][i] = df["GMV"][i] * 0.05
        if df["Service Group"][i] == "supermarket":
            df["$cashback"][i] = df["GMV"][i] * 0.05
        else:
            df["$cashback"][i] = 0
    else:
        df["$cashback"][i] = 0
```

```python
df["$cashback"]= df["$cashback"].astype("int")
```

```python
for i in range(len(df)):
    if df["$cashback"][i] > 10000:
        df["$cashback"][i] = 10000
```
